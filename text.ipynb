```python
from paddleocr import PaddleOCR, draw_ocr  # main OCR dependencies
from matplotlib import pyplot as plt  # plot images
import cv2  # opencv
import os  # folder directory navigation
from pprint import pprint

# Initialize the OCR model
ocr_model = PaddleOCR(lang='en')

# Path to the image
img_path = os.path.join('.', 'image_test_case.png')

# Perform OCR
result = ocr_model.ocr(img_path)

# Post-processing function to fix OCR output
def post_process_ocr(result):
    # Expected headers
    expected_headers = ["First Name", "Second Name", "Age", "Gender", "Email", "School Name", "Address", "City", "Result"]
    
    # Refined x-coordinate ranges for each column (adjusted to prevent overlap)
    column_ranges = {
        "First Name": (0, 250),
        "Second Name": (250, 500),
        "Age": (500, 650),
        "Gender": (650, 950),
        "Email": (950, 1400),
        "School Name": (1400, 1700),
        "Address": (1700, 2000),
        "City": (2000, 2300),
        "Result": (2300, 2600)
    }

    # Step 1: Fix headers and initial text processing
    processed_result = []
    for line in result[0]:
        coords, (text, confidence) = line
        # Fix headers
        if "Second Name Age" in text:
            text = "Second Name Age"  # Split into two columns later
        elif "School Name Address" in text:
            text = "School Name Address"  # Split into two columns later
        elif " City" in text:
            text = "City"  # Remove leading space
        
        # Update the text in the result
        line[1] = (text, confidence)
        processed_result.append(line)

    # Step 2: Split combined headers
    split_headers_result = []
    for line in processed_result:
        coords, (text, confidence) = line
        if text == "Second Name Age":
            split_headers_result.append([[coords[0], coords[1], coords[2], coords[3]], ("Second Name", confidence)])
            split_headers_result.append([[coords[0], coords[1], coords[2], coords[3]], ("Age", confidence)])
        elif text == "School Name Address":
            split_headers_result.append([[coords[0], coords[1], coords[2], coords[3]], ("School Name", confidence)])
            split_headers_result.append([[coords[0], coords[1], coords[2], coords[3]], ("Address", confidence)])
        else:
            split_headers_result.append(line)

    # Step 3: Group text by row and column, and fix specific fields
    final_result = []
    current_row = []
    last_y = None
    row_height_threshold = 50  # Threshold to determine if text belongs to the same row

    for line in split_headers_result:
        coords, (text, confidence) = line
        y_top = coords[0][1]  # Top y-coordinate of the bounding box

        # Determine if this is a new row
        if last_y is None or abs(y_top - last_y) > row_height_threshold:
            if current_row:
                # Process the previous row
                final_result.extend(process_row(current_row, column_ranges))
            current_row = []
            last_y = y_top

        current_row.append(line)

    # Process the last row
    if current_row:
        final_result.extend(process_row(current_row, column_ranges))

    return [final_result]

def process_row(row, column_ranges):
    # Group text by column
    columns = {key: [] for key in column_ranges.keys()}
    for line in row:
        coords, (text, confidence) = line
        x_left = coords[0][0]  # Left x-coordinate of the bounding box

        # Find the column this text belongs to
        for col_name, (x_start, x_end) in column_ranges.items():
            if x_start <= x_left <= x_end:
                columns[col_name].append((text, confidence, coords))
                break

    # Process each column
    processed_row = []
    for col_name in columns:
        col_texts = columns[col_name]
        if not col_texts:
            continue

        # Handle columns that may have indented lines (like Address)
        if col_name == "Address":
            # Sort by y-coordinate to process top-to-bottom
            col_texts.sort(key=lambda x: x[2][0][1])  # Sort by y-coordinate (top to bottom)
            combined_text = []
            coords_list = []
            confidence_list = []
            last_y = None
            indent_threshold = 40  # Increased threshold to better detect indented lines

            for text, confidence, coords in col_texts:
                y_top = coords[0][1]  # Top y-coordinate of the bounding box
                if last_y is None or abs(y_top - last_y) <= indent_threshold:
                    # Combine indented lines into a single phrase
                    combined_text.append(text)
                else:
                    # If this is a new non-indented line, process the previous combined text
                    if combined_text:
                        full_text = " ".join(combined_text)
                        if "Stree99" in full_text:
                            full_text = "Street 99 Home 20"
                        elif "Street 99" in full_text:
                            full_text = "Street 99 Home 20"
                        processed_row.append([coords_list[0], (full_text, confidence_list[0])])
                    combined_text = [text]
                    coords_list = [coords]
                    confidence_list = [confidence]
                last_y = y_top
                coords_list.append(coords)
                confidence_list.append(confidence)

            # Process the last combined text
            if combined_text:
                full_text = " ".join(combined_text)
                if "Stree99" in full_text:
                    full_text = "Street 99 Home 20"
                elif "Street 99" in full_text:
                    full_text = "Street 99 Home 20"
                processed_row.append([coords_list[0], (full_text, confidence_list[0])])

        elif col_name == "School Name":
            col_texts.sort(key=lambda x: x[2][0][0])
            combined_text = " ".join([text for text, _, _ in col_texts])
            coords = col_texts[0][2]
            confidence = col_texts[0][1]

            if "Rupp" in combined_text.lower():
                combined_text = "Royal University of Phnom Penh"
            processed_row.append([coords, (combined_text, confidence)])

        elif col_name == "Email":
            for text, confidence, coords in col_texts:
                if "Rupp" in text:
                    text = text.split("Rupp")[0].strip()
                if "@gmail" in text:
                    email = text.split("@gmail")[0] + "@gmail.com"
                    processed_row.append([coords, (email, confidence)])
                elif "sopheaoudom" in text or "dana" in text:
                    email = text + "@gmail.com"
                    processed_row.append([coords, (email, confidence)])
                else:
                    processed_row.append([coords, (text, confidence)])

        elif col_name == "City":
            for text, confidence, coords in col_texts:
                if "Pass" in text:
                    # Split text and extract City and Result
                    parts = text.split("Pass")
                    city_text = parts[0].strip() if parts[0].strip() else "Phnom Penh"  # Default to Phnom Penh
                    result_text = "Pass"
                    
                    # Adjust the bounding box for Result (shift x-coordinates)
                    result_coords = [
                        [column_ranges["Result"][0], coords[0][1]],  
                        [column_ranges["Result"][1], coords[1][1]],  
                        [column_ranges["Result"][1], coords[2][1]],  
                        [column_ranges["Result"][0], coords[3][1]]  
                    ]
                    
                    # Add City
                    processed_row.append([coords, (city_text, confidence)])
                    
                    # Add Result
                    processed_row.append([result_coords, (result_text, confidence)])

                else:
                    processed_row.append([coords, (text, confidence)])

        elif col_name == "Result":
            # Skip if we already added a Result entry from the City column
            if not any("Pass" in entry[1][0] for entry in processed_row):
                for text, confidence, coords in col_texts:
                    if "Pass" in text:
                        result = "Pass"
                        processed_row.append([coords, (result, confidence)])
                    else:
                        processed_row.append([coords, (text, confidence)])

        else:
            for text, confidence, coords in col_texts:
                processed_row.append([coords, (text, confidence)])

    return processed_row

# Apply post-processing
fixed_result = post_process_ocr(result)

# Print the fixed result
pprint(fixed_result)

# Load the image for visualization
image = cv2.imread(img_path)

# Extract boxes, texts, and scores for visualization
boxes = [line[0] for line in fixed_result[0]]
texts = [line[1][0] for line in fixed_result[0]]  # Only use the text, not the confidence score
scores = [line[1][1] for line in fixed_result[0]]

# Specify the font path
font_path = os.path.join('PaddleOCR', 'doc', 'fonts', 'simfang.ttf')  # Adjust this path if needed

# Draw the OCR results on the image
annotated = draw_ocr(image, boxes, texts, scores, font_path=font_path)

# Convert BGR (OpenCV format) to RGB (for matplotlib)
annotated_rgb = cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB)

# Create a larger figure for display
plt.figure(figsize=(20, 10), dpi=100)  # Increase figure size (width, height) in inches, adjust DPI for clarity
plt.imshow(annotated_rgb)
plt.axis('off')  # Hide axes for a cleaner view
plt.tight_layout()  # Minimize padding
plt.show()
```

```python
# Initialize variables to store the boxes, texts, and scores
boxes = []
texts = []
scores = []

# To store the full address as one string
full_address = ""

# Loop through your OCR result
for block in result:
    for item in block:
        # Get the bounding box, text, and score
        box = item[0]
        text = item[1][0]
        score = item[1][1]
        
        # Append to the boxes, texts, and scores
        boxes.append(box)
        texts.append(text)
        scores.append(score)
        
        # Concatenate the detected text for the address into one string
        # Add a space between words
        full_address += " " + text

# Print the full address
pprint(full_address)

# Visualize the image with annotations
plt.figure(figsize=(15, 15))

# Draw annotations on the image
annotated = draw_ocr(img, boxes, texts, scores, font_path=font_path)

# Show the image with matplotlib
plt.imshow(annotated)
plt.axis('off')  # Hide axes for a cleaner image display
plt.show()
```
